<h4>Code</h4>

```python
def min_height(root):
    """The following intuition does not work !!!
    Note: It would work for max depth (see leetcode 104)

    Counter example -  Input: root = [2,null,3,null,4,null,5,null,6]
      2
       \
        3
         \
          4
           \
            5
             \
              6
    Output: 5 (and not 1)
    """
    if not root:
        return -1  # proper use of height

    left = min_height(root.left)    # we need to account for the case
    right = min_height(root.right)  # when current node has only one child
    return 1 + min(left, right)
```

```python
def min_depth(root):
    """DFS using a bottom up recursion
    we dig until we reach a leaf (height = 0) and we add +1 for
    each node until back to the root.
    We return min(height(root))
    Note: this approach works for max_depth too - see leetcode #104)"""
    if root is None:
        return 0    # strictly speaking height(leaf) = 0 ==> depth(None) should be -1 (by convention)
    if root.left is None and root.right is None:
        return 1

    # Option 1
    if root.left is None:
        return 1 + min_depth(root.right)
    if root.right is None:
        return 1 + min_depth(root.left)
    return 1 + min(min_depth(root.left), min_depth(root.right))

    # Option 2:
    height_left = min_depth(root.left) if root.left else float("inf")
    height_right = min_depth(root.right) if root.right else float("inf")
    return 1 + min(height_left, height_right)
```

```python
def min_depth(root):
     """DFS using Top down recursion
    we dig into the tree, adding +1 at each children
    until we reach a leaf, where we return the depth we calculated so far
    we keep track of the smallest value we saw so far
    """
    def dfs(node, depth):
        nonlocal min_depth
        # base case
        if not node:
            return
        depth += 1
        if not node.left and not node.right:  # necessary (see cell #2)
            if depth < min_depth:
                min_depth = depth
            return
        left = dfs(node.left, depth)
        right = dfs(node.right, depth)
        return

    if not root:
        return 0
    min_depth = float("inf")
    dfs(root, 0)
    return min_depth
```

```python
def min_depth(root):
    """BFS using a bottom up iterative approach with a list."""
    if not root:
        return 0

    queue = [root]
    for node in queue:  
        if node:
            queue += node.left, node.right 

    height = {}  # strictly speaking it should be -1
    for node in reversed(queue):
        if node:
            if not node.left and not node.right:
                height[node] = 0
            else:
                height_left = height[node.left] if node.left else float("inf")
                height_right = height[node.right] if node.right else float("inf")
                height[node] = 1 + min(height_left, height_right)
    return 1+height[root]
```

```python
from collections import deque

def min_depth(root):
    """BFS using a bottom up iterative approach  with a deque."""
    if not root:
        return 0
    
    queue = deque()
    queue.append(root)
    traversal = deque()
    while queue:
        node = queue.popleft()
        if node:
            traversal.appendleft(node)
            queue += node.left, node.right  # last nodes --> full layer of "None"

    height = {}
    while traversal:
        node = traversal.popleft()
        if node:
            if not node.left and not node.right:
                height[node] = 0
            else:
                height_left = height[node.left] if node.left else float("inf")
                height_right = height[node.right] if node.right else float("inf")
                height[node] = 1 + min(height_left, height_right)
    return 1+height[root]
```
