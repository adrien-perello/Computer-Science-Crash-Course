<h4>Code</h4>

```python
def get_intersection_node(headA, headB):
    """Using Hash Table
    Time Complexity:  O(n)
    Space complexity: O(n)
    """
    cache = set()
    while headA:
        cache.add(headA)
        headA = headA.next
    while headB:
        if headB in cache:
            return headB
        headB = headB.next
    return None
```

<h4>Code</h4>

```python
def get_intersection_node(headA, headB):
    """Using Two Pointers.

    The idea here is to compensate the prefix length difference.
    Start a pointer at head A and B and chain them once they reach the end to
    head B and A. The reason is that the difference in prefix length will be
    offset by the time the pointers hit the intersection for the second time.

    A = 1,3,5,7,9,8
    B = 2,4,9,8

    A+B = 1,3,5,7,9,8,2,4 | 9,8
    B+A = 2,4,9,8,1,3,5,7 | 9,8

    Time Complexity O(n) = O(A1+B1+C)
    Space complexity O(1)
    """
    a, b = headA, headB
    while a != b:
        a = headB if a is None else a.next
        b = headA if b is None else b.next
    return a
```
