```python
import numpy as np
import xarray as xr
from fractions import Fraction
from dataclasses import dataclass
import plotly.graph_objects as go
```

```python
def nparange(start, end, step, dtype=np.float32):
    """Modified np.arange()
        - improve float precision
        - includes endpoint

    Args:
        start, end, step: float (end is included in array)

    Returns:
        ndarray
    """
    frac = Fraction("{0:.20f}".format(step))
    delta = frac.numerator   # new step
    zoom = frac.denominator  # zoom to avoid float precision error
    if zoom > 1e20:
        print("WARNING: Ensure that step is a rational number"
            " to avoid floating errors")
    res = np.arange(start * zoom, end * zoom + delta, delta) / zoom
    return res.astype(dtype, casting='same_kind') 
```

```python
Ux = nparange(-4, 4, 1)
Uy = nparange(-5, 5, 2)
Vx = nparange(0,10,2)
Vy = nparange(1,11, 1.5)
```
