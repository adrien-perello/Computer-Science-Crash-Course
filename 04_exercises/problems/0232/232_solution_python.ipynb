<h4>Code</h4>

```python
class MyQueue:
    """Priority to Pop."""

    def __init__(self):
        self.s1 = []
        self.s2 = []

    def push(self, x):  # 
        """Push newest element to the bottom of the stack (i.e. s1).
        Time Complexity O(n)
        Space Complexity O(n)
        
        Note: Each element, with the exception of the newly arrived,
        is pushed and popped twice.
        The last inserted element is popped and pushed once.
        Therefore this gives 4n+2 operations where n is the queue size.
        """
        while self.s1:
            self.s2.append(self.s1.pop())  # transfer all s1 elements to stack s2
        self.s1.append(x)                  # then, newly arrived element is pushed at the bottom of s1
        while self.s2:
            self.s1.append(self.s2.pop())  # and all s2 elements are popped and pushed to s1

    def pop(self):
        """
        Time Complexity O(1)
        Space Complexity O(1)
        """
        if self.s1:
            return self.s1.pop()

    def peek(self):
        """
        Time Complexity O(1)
        Space Complexity O(1)
        """
        return self.s1[-1]

    def empty(self):
        """
        Time Complexity O(1)
        Space Complexity O(1)
        """
        return self.s1 == []
```

<h4>Follow up:</h4>
<p>Can you implement the queue such that each operation is <strong><a href="https://en.wikipedia.org/wiki/Amortized_analysis" target="_blank">amortized</a></strong> <code>O(1)</code> time complexity? In other words, performing <code>n</code> operations will take overall <code>O(n)</code> time even if one of those operations may take longer.</p>

<h4>Code</h4>

```python
class MyQueue:
    """Priority to Push."""

    def __init__(self):
        self.s1 = []
        self.s2 = []
        self.front = None

    def push(self, x):
        """
        Time Complexity O(1)
        Space Complexity O(n)
        """
        if self.s1 == []:
            self.front = x
        self.s1.append(x)

    def pop(self):
        """
        Time Complexity Amortized: O(1) // Worst Case: O(n)
        Space Complexity O(n)
        
        Note: In the worst case scenario when stack s2 is empty,
        the algorithm pops n elements from stack s1 and pushes n elements to s2,
        where n is the queue size. This gives 2n operations, which is O(n).
        But when stack s2 is not empty the algorithm has O(1) time complexity.
        
        Amortized analysis gives the average performance (over time) of each
        operation in the worst case. The basic idea is that a worst case
        operation can alter the state in such a way that the worst case can't
        occur again for a long time, thus amortizing its cost.
        """
        if self.s2 == []:
            while self.s1:
                self.s2.append(self.s1.pop())
        return self.s2.pop()

    def peek(self):
        """
        Time Complexity O(1)
        Space Complexity O(1)
        """
        if self.s2:
            return self.s2[-1]
        return self.front

    def empty(self):
        """
        Time Complexity O(1)
        Space Complexity O(1)
        """
        return self.s1 == [] and self.s2 == [] 
```
